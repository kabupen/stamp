# コスト関数

```python
import numpy as np
import matplotlib.pyplot as plt

# 格子点座標
w_x = np.arange(-10, 10+1, 0.1)
w_y = np.arange(-10, 10+1, 0.1)

w_xx, w_yy = np.meshgrid(w_x, w_y)

J = []
J_L2 = []
for idx_x in range(0, w_x.shape[0]):
    for idx_y in range(0, w_y.shape[0]):
        
        weight = np.array([w_xx[idx_x][idx_y], w_yy[idx_x][idx_y]])

        N = 5
        np.random.seed(1)
        X = np.random.randn(N, 2)
        y = np.logical_xor(X[:,0]>0, X[:, 1]>0)
        y = np.where(y, 1, -1)
        
        error = y - (X*weight).sum(axis=1)
        J.append(0.5 * (error**2).sum())
        
        J_L2.append(0.5 * (error**2).sum()+ (weight**2).sum())

J = np.array(J)
J = J.reshape(w_xx.shape)

J_L2 = np.array(J_L2)
J_L2 = J_L2.reshape(w_xx.shape)

fig, ax = plt.subplots()

plt.xlim(-10, 10)
plt.ylim(-10, 10)

#plt.contourf(w_xx, w_yy, J)
cntr = ax.contour(w_xx, w_yy, J, levels=20)
#ax.clabel(cntr)
ax.axhline(y=0., linewidth=1, color='k', linestyle='-')
ax.axvline(x=0., linewidth=1, color='k', linestyle='-')

plt.show()

fig, ax = plt.subplots()
plt.xlim(-10, 10)
plt.ylim(-10, 10)

#plt.contourf(w_xx, w_yy, J)
cntr = ax.contour(w_xx, w_yy, J_L2, levels=20)
#ax.clabel(cntr)
ax.axhline(y=0., linewidth=1, color='k', linestyle='-')
ax.axvline(x=0., linewidth=1, color='k', linestyle='-')

plt.show()
```
