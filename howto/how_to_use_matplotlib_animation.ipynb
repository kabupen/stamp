```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.animation as animation
```

```python
%matplotlib nbagg
np.random.seed(123)
x = np.random.randn(100)

absmax_x = np.ceil(np.max(np.abs(x)))
xmin, xmax = -absmax_x, absmax_x

def animate(nframe):
    plt.cla()
    plt.hist(x[:nframe], bins=20)
    plt.xlim(xmin , xmax)
    plt.ylim(0, 50)
    
fig = plt.figure()
anim = animation.FuncAnimation(fig, animate, np.arange(1, 100+1))
plt.show()
```
